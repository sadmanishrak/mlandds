```python
import pandas as pd
car_sales = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv")
car_sales
```

```python
# Doesn't work because Price is a string type
car_sales["Price"].plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "")
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales["Sale Date"] = pd.date_range("17/5/2020", periods=len(car_sales))
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales")
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(kind="scatter", x = "Odometer (KM)", y="Price")
```

```python
heart_disease = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
```

```python
heart_disease.head(10)
```

```python
heart_disease.plot.hist(bins=50, by="age")
```

```python
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30))
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(),
           linestyle="--");

```

```python
over_50 = heart_disease[heart_disease["age"] > 50]

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"],
                     cmap="winter")
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(),
           linestyle="--",
           color="red");
```
